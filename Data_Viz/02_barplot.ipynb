```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

# draw a barplot
sns.barplot(x="species", y= "sepal_width", data=phool)
plt.show()
```
